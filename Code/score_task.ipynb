```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("dataset.csv")
```

```python
data.head()
```

```python
data['score'] = data['score'].fillna(0)
```

```python
# Subject Duplicates
dups_subject = data.pivot_table(index=['subject'], aggfunc='size', )
print(dups_subject)
```

```python
subject_result_dataframe = pd.DataFrame(dups_subject)
subject_result_dataframe = subject_result_dataframe.reset_index()
subject_result_dataframe.columns = ['subject', 'subject_count']
subject_result_dataframe
data = pd.merge(data, subject_result_dataframe, on='subject')
data
```

```python
# Object duplicates
dups_object = data.pivot_table(index=['object'], aggfunc='size')
print (dups_object)
```

```python
object_result_dataframe = pd.DataFrame(dups_object)
object_result_dataframe = object_result_dataframe.reset_index()
object_result_dataframe.columns = ['object', 'object_count']
object_result_dataframe
data = pd.merge(data, object_result_dataframe, on='object')
data
```
